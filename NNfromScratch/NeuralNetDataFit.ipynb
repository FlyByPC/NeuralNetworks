```python
import tensorflow as tf
import numpy as np
from math import log2
```

```python
x = np.linspace(0, 40, 20)
y = np.array([i + np.random.normal(0, 1, 1) for i in x])
#yn = np.array([i*math.log2(i) + np.random.normal(0, 1, 1) for i in x])
```

```python
x.size
```

```python
xx = np.linspace(0, 40, 20)
yy = np.array([0.3*(i + np.random.normal(0, 3, 1)) * (i + np.random.normal(0, 3, 1)) for i in x])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(1)
])
```

```python
modelLinear = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
model.fit(x, y, validation_split=0.05, epochs=1000, shuffle=True, verbose=2)
```

```python
modelLinear.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
modelLinear.fit(x, y, validation_split=0.05, epochs=1000, shuffle=True, verbose=2)
```

```python
predictions = []
predictionsLinear = []
for i in np.linspace(0, 40, 100):
    predictions.append(model.predict([i], verbose=1))
    predictionsLinear.append(modelLinear.predict([i], verbose=1))
```
